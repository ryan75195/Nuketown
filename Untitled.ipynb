# HSV Colour Space

Takes the RGB map and converts it to the HSV colour space for colour detection

```python
HSV = cv2.cvtColor(rgb_map, cv2.COLOR_BGR2HSV)
plt.figure(figsize = (20,10))
plt.imshow(HSV)
```

# Colour Detection - Extracting Objects

using the HSV map we create a mask using the given hsv ranges for each colour to isolate each colour/object on the map. with this object isolated we can then use our getCoords() function to extract the position of each pixel in the mask which is not black(ie the object) and add this entry to our dictionary.

# Walls

```python
lower_hsv = np.array([53,0,255])
higher_hsv = np.array([63,255,255])
mask = cv2.inRange(HSV, lower_hsv, higher_hsv)
result = cv2.bitwise_and(bgr_map, bgr_map, mask=mask)
plt.figure(figsize = (20,10))
plt.imshow(result)

coordinates['walls'] = getCoords(result)

```

# Writing Results to File

writes each entry in our dictionary to a file to allow us to impliment them in our game

```python
f = open("Coordinates.txt", "w")

for obj in coordinates:
    f.write("{} \n".format(coordinates[obj]))
f.close()
```
